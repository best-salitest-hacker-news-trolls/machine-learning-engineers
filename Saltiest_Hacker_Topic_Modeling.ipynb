```
df = pd.read_json('https://raw.githubusercontent.com/best-salitest-hacker-news-trolls/machine-learning-engineers/master/10000_predictions_extra.json')
```

```
import numpy as np
import gensim
import os
import re

from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from gensim import corpora

from gensim.models.ldamulticore import LdaMulticore

import pandas as pd
```

```
STOPWORDS = set(STOPWORDS).union(set(['think', 'looks', 'hn', 'link', 'like']))

def tokenize(text):
    return [token for token in simple_preprocess(text) if token not in STOPWORDS]
```

```
df['tokens'] = df['text'].apply(tokenize)
```

```
# creating dictionary 

id2word = corpora.Dictionary(df.tokens.to_list())
```

```
len(id2word.keys())
```

```
# Let's remove extreme values from the dataset
# same like the parameters 'min' and 'max'

id2word.filter_extremes(no_below=5, no_above=0.95)
```

```
# this causes a decrease in the words

len(id2word.keys())
```

```
corpus = df.tokens.apply(id2word.doc2bow)
```

```
lda = LdaMulticore(corpus=corpus,
                   id2word=id2word,
                   random_state=42,
                   num_topics = 20,
                   chunksize=100,
                   iterations=400,
                   passes=5,
                   workers=2
                  )
```

```
lda.print_topics()
```

```
words = [re.findall(r'"([^"]*)"',t[1]) for t in lda.print_topics()]
```

```
topics = [' '.join(t[0:5]) for t in words]
```

```
for id, t in enumerate(topics): 
    print(f"------ Topic {id} ------")
    print(t, end="\n\n")
```
